```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definir la funcin original
def f(x):
    return np.exp(x)

# Calcular la serie de Taylor hasta el orden n en torno a x = a
def taylor_expansion(x, a, n):
    return sum([(f(a) * (x - a)**i) / math.factorial(i) for i in range(n+1)])

# Punto de expansin
a = 0  # Puedes cambiar este valor

# Rango de valores para graficar alrededor de a
x = np.linspace(a - 2, a + 2, 100)
y_real = f(x)

# Niveles de aproximacin a visualizar
#orders = [1, 2, 4, 6, 10]
orders = [1, 2]

plt.figure(figsize=(8, 6))

# Graficar la funcin original
plt.plot(x, y_real, label=r'$e^x$ (Funcin real)', linewidth=2, color='black')

# Graficar aproximaciones de Taylor de distintos rdenes
for n in orders:
    y_taylor = np.array([taylor_expansion(xi, a, n) for xi in x])
    plt.plot(x, y_taylor, linestyle='dashed', label=f'Taylor Orden {n} en a={a}')

# Configuracin del grfico
plt.axhline(y=0, color='gray', linewidth=0.5)
plt.axvline(x=a, color='red', linestyle='dotted', label=f'Punto de expansin a={a}')
plt.legend()
plt.title(f"Aproximacin de $e^x$ con la Serie de Taylor en $a={a}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
